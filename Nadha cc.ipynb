```python
def add(x,y):
  return x+y
def sub(x,y):
  return x-y
def multiplication(x,y):
  return x*y
def div(x,y):
  return x//y
def exp(x,y):
  return x**y
print("calculator")
print("select operation")
print("1.addition")
print("2.substraction")
print("3.multiplication")
print("4.division")
print("5.exponential")
choice=int(input("enter 1/2/3/4/5:"))
if choice in (1,2,3,4,5):
  x=float(input("enter a number"))
  y=float(input("enter a number"))
  if choice==1:
    print(x,"+",y,"=",add(x,y))
  elif choice==2:
    print(x,"-",y,"=",sub(x,y))
  elif choice==3:
    print(x,"*",y,"=",multiplication(x,y))
  elif choice==4:
    print(x,"//",y,"=",div(x,y))
  elif choice==5:
    print(x,"^",y,"=",exp(x,y))
```
